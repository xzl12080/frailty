```python
from collections import Counter
import numpy as np
data = pd.read_csv(r'C:/Users/10634/Desktop/recluster_all_2000/54230第一个班型的课程时长及时间.csv')
data = data.iloc[:,1:]
data.columns = ['question','class']
raw_dic = {}
cluster_center = {} 
c = dict(Counter(data['class']))
for class_id,num in c.items():
    key = f'class{class_id}'
    value = list(data[data['class'] == class_id]['question'])
    raw_dic[key] = value
    s = SentenceTransformer('shibing624/text2vec-base-chinese')
    center_vec = np.mean(s.encode(value),axis=0)
    cluster_center[class_id] = center_vec
new_dic = {}
for qu in list(data['question']):
    qu_vec = s.encode(qu)
    sim_list = []
    for class_id,center in cluster_center.items():
        sim = cosine_similarity(qu_vec.reshape(1,-1),center.reshape(1,-1))
        sim_list.append([class_id,sim])
    slist = []
    idlist = []
    for i in sim_list:
        idlist.append(i[0])
        slist.append(i[1])
    max_sim = max(slist)
    max_id = idlist[slist.index(max_sim)]
    if max_id not in new_dic.keys():
        new_dic[max_id] = [qu]
    else:
        new_dic[max_id].append(qu)
new_dic
```
