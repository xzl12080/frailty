```python
path = r'C:\Users\10634\Desktop\bad case'
for root,name,files in os.walk(path):
    for file in files:
        print(file)
```

```python
df_list = []
for file in files:
    filepath = root + '\\' + file
    data = pd.read_excel(filepath)
    data = data.iloc[:,0:-1]
    data.columns = data.loc[0]
    data = data.iloc[1:,]
    data = data.fillna('0')

    fn = data[(data['标注结果'] == '1') & (data['模型预测结果'] == '否')]
    fp = data[(data['标注结果'] == '0') & (data['模型预测结果'] == '是')]
    df = pd.concat([fn,fp],axis=0)
    index = [file[:-5]]*len(list(df.index))
    df.index = index
    df_list.append(df)
badcase = pd.concat(df_list,axis=0)
badcase
```

```python
badcase.to_csv('./badcase2.csv',encoding='utf-8-sig')
```
