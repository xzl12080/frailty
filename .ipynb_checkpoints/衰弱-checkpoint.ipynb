```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from tqdm import tqdm
path = r"C:\Users\10634\Desktop\frailty\data.xlsx"

drop_feature = ['衰弱评估','其他疾病','手表','缝纫机','自行车']
# 原始数据
df = pd.read_excel(path)
for i in drop_feature:
    df.drop(i,axis=1,inplace=True)

# 分类变量'Frail诊断','Fried诊断'
df_class = pd.read_excel(path,sheet_name='分类变量')
for i in drop_feature:
    df_class.drop(i,axis=1,inplace=True)
class_feature = list(df_class.columns)

# 数值变量
df_numeric = pd.read_excel(path,sheet_name='数值变量')
numeric_feature = list(df_numeric.columns)

threshold = 0.01
variances = df.var()
# 找出低方差特征
low_variance_columns = variances[variances < threshold].index

# # 打印低方差特征
# print("低方差特征:")
# print(low_variance_columns)

# 删除低方差特征
df = df.drop(columns=low_variance_columns)

df.dropna(thresh=int(len(df.index) / 3),axis=1,inplace=True) # 删除缺失值超过三分之二的特征

class_feature = [i for i in df.columns if i in class_feature]
numeric_feature = [i for i in df.columns if i in numeric_feature]
df = pd.concat([df[class_feature],df[numeric_feature]],axis=1)
# df['衰弱'] = df['衰弱'].astype('category')
```

```python
df
```

```python
df.fillna(value=888,inplace=True)
error_feature = ['Fried诊断','职业','婚姻','合作医疗','糖尿病','疾病数目','手术史','外伤史','输血史','30岁前锻炼','目前锻炼','不良事件']
# 删除异常样本
del_index = []
for i in error_feature:
    c = Counter(df[i])
    c = dict(c)
    for key in c.keys():
        if c[key] < 10 and key != 888:
            del_index.append(df[df[i] == key].index)
del_index = list(set([item for i in del_index for item in i]))
for i in del_index:
    df = df[df.index != i]
df.replace(888,np.nan,inplace=True)

for i in class_feature:
#     df[i] = df[i].astype('int')
    df[i] = df[i].astype('category')
```

## 重采样前

### 汉族

```python
#评估器为随机森林的递归特征选择法即包装法
from sklearn.ensemble import RandomForestClassifier as RFC_  #分类器
from sklearn.feature_selection import RFE  #递归特征消除法


rf=RFC_(random_state=42)
rf=rf.fit(h_x,h_y)
h_score = []
for i in tqdm(range(1,len(df_h.columns)+1),desc='递归法选择特征:'):
    selector_RFE = RFE(rf,
                    n_features_to_select=i,
                    step=1
                    ).fit(h_xtrain, h_ytrain)
    h_score.append(selector_RFE.score(h_xtest,h_ytest))

plt.figure(figsize=(9,5))

plt.plot(np.arange(len(df_h.columns)),h_score)
#over_x.columns[selector_RFE.get_support()]
# hx.columns[selector_RFE.get_support()]
from sklearn.metrics import precision_score,recall_score,f1_score,roc_auc_score
rfe = RFE(
    rf,
    n_features_to_select=h_score.index(max(h_score))+1,
    step=1
    ).fit(h_xtrain,h_ytrain)
h_x.columns[rfe.get_support()]
```

### 蒙族

```python
#评估器为随机森林的递归特征选择法即包装法
from sklearn.ensemble import RandomForestClassifier as RFC_  #分类器
from sklearn.feature_selection import RFE  #递归特征消除法


rfm = RFC_(random_state=42)
rfm = rfm.fit(m_x,m_y)
m_score = []
for i in range(1,len(df_m.columns)+1):
    selector_RFE = RFE(rfm,
                    n_features_to_select=i,
                    step=1
                    ).fit(m_xtrain, m_ytrain)
    m_score.append(selector_RFE.score(m_xtest,m_ytest))

plt.figure(figsize=(9,5))

plt.plot(np.arange(len(df_m.columns)),m_score)
#over_x.columns[selector_RFE.get_support()]
# hx.columns[selector_RFE.get_support()]
```

# 聚类

## K均值聚类

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
hx_normalized = scaler.fit_transform(h_x[numeric_feature])
df_h[numeric_feature] = hx_normalized
df_h
```

### 汉族

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# 使用K均值聚类
score_list = []
for i in tqdm(range(2,11)):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(df_h)
    score = silhouette_score(df_h, labels)
    # 评估聚类效果（使用轮廓系数）
    score_list.append(score)

plt.figure(figsize=(9,7))
plt.plot(np.arange(2,11),score_list)
plt.show()

```
